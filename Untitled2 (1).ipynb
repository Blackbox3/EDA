```python

import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python

# Load dataset
def load_dataset(file_path):
  return pd.read_csv(file_path)


df = load_dataset("/content/drive/MyDrive/archive (2)/CarPrice_Assignment.csv")
df.head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the car price dataset

# Countplot for 'Brand'
sns.countplot(data=df, x='CarName')
plt.title('Count of Cars by CarName')
plt.xlabel('CarName')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the car price dataset

# Pie chart for 'Fuel Type'
df['fueltype'].value_counts().plot(kind='pie', autopct='%0.2f%%')
plt.title('Distribution of Cars by Fuel Type')
plt.ylabel('')  # Remove the ylabel
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the car price dataset

# Replace 'Age' with the numerical column you want to visualize
plt.hist(df['price'])  # Example: 'Price' is a numerical column in the car price dataset
plt.title('Distribution of Car Prices')
plt.xlabel('price')
plt.ylabel('horsepower')
plt.show()
```
